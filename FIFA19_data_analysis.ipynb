```python
#fifa19 데이터 불러오기
import numpy as np
import pandas as pd
df = pd.read_csv(r'C:\Users\USER\Desktop\leemingo\study\fifa data\data.csv')
```

```python
#분석에 필요한 column 취합
columns = ['Name', 'Age', 'Nationality', 'Club', 'Overall', 'Potential', 'Preferred Foot',
           'Crossing', 'Finishing','HeadingAccuracy', 'ShortPassing',
           'Volleys', 'Dribbling','Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes'
          ]
```

```python
df = pd.DataFrame(df, columns = columns)
```

```python
#플레이어1의 선수정보 입력
class NameException(Exception):
    def __init__(self, msg):
        self._message = msg
    def player_insert():
        team = input('Input your team name:')
        
        if team.isdigit() == True:
            raise NameException(' 팀 이름을 입력하세요.')
        else:
            return team
if __name__ == '__main__':
    
    try:
        p1 = NameException.player_insert()
        
    except NameException as n:
        print(n.args[0])
    else:
        print(p1)

        p1_GK_name = input('input your GK:')
        p1_LB_name = input('input your LB:')
        p1_LCB_name = input('input your LCB:')
        p1_RCB_name = input('input your RCB:')
        p1_RB_name = input('input your RB:')
        p1_CDM_name = input('input your CDM:')
        p1_CM_name = input('input your CM:')
        p1_CAM_name = input('input your CAM:')
        p1_LW_name = input('input your LW:')
        p1_RW_name = input('input your RW:')
        p1_ST_name = input('input your ST:')

```

```python
#플레이어1의 선수들을 dataFrame에 저장
gk = df[(df.Name == p1_GK_name)&(df.Club == p1)]
lb = df[(df.Name == p1_LB_name)&(df.Club == p1)]
lcb = df[(df.Name == p1_LCB_name)&(df.Club == p1)]
rcb = df[(df.Name == p1_RCB_name)&(df.Club == p1)]
rb = df[(df.Name == p1_RB_name)&(df.Club == p1)]
cdm = df[(df.Name == p1_CDM_name)&(df.Club == p1)]
cm = df[(df.Name == p1_CM_name)&(df.Club == p1)]
cam = df[(df.Name == p1_CAM_name)&(df.Club == p1)]
lw = df[(df.Name == p1_LW_name)&(df.Club == p1)]
rw = df[(df.Name == p1_RW_name)&(df.Club == p1)]
st = df[(df.Name == p1_ST_name)&(df.Club == p1)]
my_team = pd.concat([gk,lb, lcb, rcb, rb, cdm, cm, cam, lw, rw, st], axis = 0, ignore_index = True)
my_team[(my_team.Club == 'FC Barcelona')]
```

```python
#플레이어 1의 팀 평균값 구하기

def team_average(row):
    team_avg = round(float((my_team.loc[0, row] + my_team.loc[1, row] + my_team.loc[2, row] + my_team.loc[3, row] +
                          my_team.loc[4, row] + my_team.loc[5, row] + my_team.loc[6, row] + my_team.loc[7, row] +
                          my_team.loc[8, row] + my_team.loc[9, row] + my_team.loc[10, row])/11),2)
    return team_avg
my_team.loc[11] = np.nan
for stats in my_team.columns:
    if type(my_team.loc[0,stats]) == float:
        my_team.loc[11, stats] = team_average(stats)
    else:
        my_team.loc[11, stats] = '-'

my_team
```

```python
#팀 평균 데이터를 새로운 DataFrame에 저장
my_team_avg = my_team.loc[11]
your_team_avg = your_team.loc[11]

comparison = pd.DataFrame({'Team1': my_team_avg, 'Team2': your_team_avg})

comparison
```

```python
#시각화를 위해 non-numeric data row 삭제 및 상대팀과의 능력치 비교
for index in comparison.index:
    if type(comparison.loc[index, 'Team1']) == str:
        comparison.drop([index], inplace = True)
comparison['Comparison'] = comparison['Team1'] - comparison['Team2']
comparison
```

```python
#상대 팀과 비교하여 강약점 추출

print('[Strong Stats]')
for index in comparison.index:
    if comparison.loc[index, 'Comparison'] > 3:
        print('\n',index, end = '')
print('')
print('\n[Weak Stats]')
for index in comparison.index:
    if comparison.loc[index, 'Comparison'] < -3:
        print('\n', index, end = '')
```

```python
#특정 주요 데이터 이용하여 플레이 방향 제시

print('Overall')
if comparison.loc['Overall', 'Team1'] > comparison.loc['Overall', 'Team2']:
    print('상대팀보다 평균적으로 능력치가 높습니다.')
else: print('상대팀이 평균적으로 능력치가 높습니다.')

print()
print('Strength')
    
if comparison.loc['Strength', 'Team1'] > comparison.loc['Strength', 'Team2']:
    print('상대팀보다 체격적으로 우위에 있습니다. \n몸싸움을 적극적으로 시도할 필요가 있습니다.')
else: print('상대팀이 체격적으로 우위에 있습니다. \n몸싸움을 가급적 피해서 플레이 하십시오.')

print()
print('ShortPassing')

if comparison.loc['ShortPassing', 'Team1'] > comparison.loc['ShortPassing', 'Team2']:
    print('상대팀보다 짧은 패스에 강점이 있습니다. \n짧은 패스를 이용해서 공격하십시오.')
else: print('상대팀이 짧은 패스에 강점이 있습니다. \n짧은 패스를 이용한 공격에 대비하십시오.')

print()
print('Agility')

if comparison.loc['Agility', 'Team1'] > comparison.loc['Agility', 'Team2']:
    print('상대팀보다 속력에 강점이 있습니다. \n속공을 주로 이용하여 공격하십시오.')
else: print('상대팀이 속력에 강점이 있습니다. \n속공을 이용한 공격에 대비하십시오.')

print()
print('Stamina')

if comparison.loc['Stamina', 'Team1'] > comparison.loc['Stamina', 'Team2']:
    print('상대팀보다 체력에 강점이 있습니다. \n체력이 떨어지는 후반에 기회를 만드십시오.')
else: print('상대팀이 체력에 강점이 있습니다. \n교체선수를 활용하여 후반전에 대비하십시오.')
```

```python
import time
class CheckTime:
    def __init__(self, f):
        print('Player Recommendation for your team')
        self.func = f
    def __call__(self, *args, **kwargs):
        start = time.time()
        self.func(*args, **kwargs)
        end = time.time()
        print('실행시간:', end - start)

@CheckTime
def player_recommendation():
    top_player = df[df.Overall >= 85]

    weak_stats = []
    for index in comparison.index:
        if comparison.loc[index, 'Comparison'] < -3:
            weak_stats.append(index)

    top_player['count'] = np.nan
    for index in top_player.index:
        count = 0
        for ws in weak_stats:
            if comparison.loc[ws, 'Team1'] <  top_player.loc[index, ws]:
                count += 1
        top_player.loc[index, ['count']] = count
        if int(top_player.loc[index, ['count']]) > len(weak_stats) -2:
            print(top_player.loc[index, 'Name'])

player_recommendation()

        
```
